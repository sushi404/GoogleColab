```python
import MeCab
import pandas as pd

def tokenize(text):
    tagger = MeCab.Tagger()
    tagger.parse('')

    tokens = []
    node = tagger.parseToNode(text)

    while node:
        surface = node.surface
        if surface != '':
            tokens.append(surface)
        node = node.next

    return tokens

# テキストの回数を入力
num_texts = int(input("テキストの回数を入力してください: "))

texts = []
for i in range(num_texts):
    text = input(f"テキスト{i+1}を入力してください: ")
    texts.append(text)

# 形態素解析を実行し、結果をデータフレームに格納
results = []
for i, text in enumerate(texts):
    tokens = tokenize(text)
    results.append({"テキスト": f"テキスト{i+1}", "形態素": tokens})

# 結果を表形式で表示
df = pd.DataFrame(results)
print(df)
```

```python
import MeCab
import pandas as pd

def tokenize(text):
    # MeCabのインスタンスを作成
    tagger = MeCab.Tagger()
    tagger.parse('')  # 文字列を解析するための準備を行う

    tokens = []
    node = tagger.parseToNode(text)

    while node:
        # 形態素の表層形を取得
        surface = node.surface
        if surface != '':
            tokens.append(surface)
        node = node.next

    return tokens

# 入力回数の取得
count = int(input("形態素解析を行うテキストの回数を入力してください: "))

# テキストと形態素解析結果を保持するリスト
texts = []
results = []

# テキストと形態素解析結果の取得
for i in range(count):
    text = input("テキストを入力してください: ")
    texts.append(text)
    result = tokenize(text)
    results.append(result)

# データフレームの作成
data = {'テキスト': texts, '形態素解析結果': results}
df = pd.DataFrame(data)

# 表形式で表示
print(df)
```
